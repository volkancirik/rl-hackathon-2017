```python
import gym
from gym import wrappers
import os
```

```python
def rmdir(top):
    for root, dirs, files in os.walk(top, topdown=False):
        for name in files:
            os.remove(os.path.join(root, name))
        for name in dirs:
            os.rmdir(os.path.join(root, name))
```

```python
def video(fname, mimetype):
    from IPython.display import HTML
    video_encoded = open(fname, "rb").read().encode("base64")
    video_tag = '<video controls alt="test" src="data:video/{0};base64,{1}">'.format(mimetype, video_encoded)
    return HTML(data=video_tag)

```

```python
def run_ol():
    env = gym.make('ObjectLocalization-v0')
    EXP_FOLDER = '/tmp/ol-experiment-0/'
    rmdir(EXP_FOLDER)
    env = wrappers.Monitor(env, EXP_FOLDER)
    #env = gym.wrappers.Monitor(env, directory)

    for agent in range(1):
        env.reset()
        done = False
        for t in range(200):
            action = env.action_space.sample()
            observation, reward, done, info = env.step(action)
            if done:
                print("Agent {} Episode finished after {} timesteps with reward {}".format(agent, t+1, reward))
                break
        if reward > 0:
            env.close()
            break
    env.close()
    return EXP_FOLDER

EXP_FOLDER = run_ol()
for file in os.listdir(EXP_FOLDER):
    if file.endswith(".mp4"):
        video_file = file
        print video_file
video_file = EXP_FOLDER + video_file
print "showing environment record at:", video_file   
video(video_file, "mp4")
```
